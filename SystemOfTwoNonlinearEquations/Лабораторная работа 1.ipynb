```python
#Подключение модулей и библиотек
import numpy as np
import matplotlib.pyplot as plt
import math
from sympy import *
```

### Первая часть
Визуализация

```python
x, y = np.mgrid[-10:10:100j,
                -10:10:100j]
ReF = np.cosh(x)*np.cos(y)-x-1
ImF = np.sinh(x)*np.sin(y)-y

fig, ax = plt.subplots()

ax.contour(ReF,colors = 'r')
ax.contour(ImF,colors = 'c')

ax.grid()

fig.set_figwidth(8)
fig.set_figheight(8)
plt.show()
```
